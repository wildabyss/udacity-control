Q2. What is the effect of the PID according to the plots, how each part of the PID affects the control command?

Proportional gain increases the rate of response, but introduces overshoots and oscillations.
Integral gain zeros out the steady-state error, but can introduce oscillations as well. 
Derivative gain decreases the settling time, but can destabilize the system when over-gained.

Q3. How would you design a way to automatically tune the PID parameters?

One way is to design the gains targeting rise-time, settling time and steady-state error assuming a step input. The vehicle plant model is linear and we can introduce linear disturbances.

Q4. PID controldel free controller, i.e. it does not use a model of the car. Could you explain the pros and cons of this type of controller?

Model-free means that the PID controller as tuned is generally robust to model variation within the gain and phase margins. However, lack of model also makes performance suffer since the gains are agnostic to any learning on the model.

Q5. (Optional) What would you do to improve the PID controller?

We can add feedforward to speed up the response, since we're given a trajectory over a horizon.
The sampling rate of the controller is very low. It would benefit by speeding up the rate of localization and control.
